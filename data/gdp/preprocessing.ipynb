```python
import pandas as pd
import country_converter as coco
```

```python
gdp = pd.read_excel('gdp.xls')
gdp
```

```python
gdp.dropna(inplace=True)
```

```python
gdp
```

```python
gdp.rename(columns={'GDP, current prices (Billions of U.S. dollars)': 'country'}, inplace=True)
```

```python
gdp
```

```python
gdp_melted = gdp.melt(id_vars=['country'], value_vars=[col for col in gdp.columns if col != 'country'], value_name='gdp', var_name='year')
```

```python
gdp_melted = gdp_melted[(gdp_melted['year'] >= 2018) & (gdp_melted['year'] <= 2023)]
gdp_melted
```

```python
cc = coco.CountryConverter()
gdp_melted['country'] = gdp_melted['country'].apply(lambda x: cc.convert(x, to='name_short'))
```

```python
gdp_melted[gdp_melted['country'] == 'United States']
```

```python
gdp_melted.to_csv('gdp.csv')
```
