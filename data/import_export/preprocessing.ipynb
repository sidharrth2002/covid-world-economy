```python
import pandas as pd
import country_converter as coco
```

```python
imports = pd.read_excel('gdp_raw.xlsx', sheet_name='Imports')
exports = pd.read_excel('gdp_raw.xlsx', sheet_name='Exports')
```

```python
imports.dropna(inplace=True)
exports.dropna(inplace=True)
```

```python
cols_2021 = [col for col in imports.columns if '2021' in col]
cols_2021
```

```python
cols_2019 = [col for col in imports.columns if '2019' in col]
cols_2020 = [col for col in imports.columns if '2020' in col]
cols_2021 = [col for col in imports.columns if '2021' in col]
imports['2019'] = imports[cols_2019].sum(axis=1)
imports['2020'] = imports[cols_2020].sum(axis=1)
imports['2021'] = imports[cols_2021].sum(axis=1)

cols_2019 = [col for col in exports.columns if '2019' in col]
cols_2020 = [col for col in exports.columns if '2020' in col]
cols_2021 = [col for col in exports.columns if '2021' in col]
exports['2019'] = exports[cols_2019].sum(axis=1)
exports['2020'] = exports[cols_2020].sum(axis=1)
exports['2021'] = exports[cols_2021].sum(axis=1)
```
