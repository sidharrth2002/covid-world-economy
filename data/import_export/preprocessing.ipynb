```python
import pandas as pd
import country_converter as coco
```

```python
import_export = pd.read_csv('import-export_raw.csv')
import_export.head()
```

```python
import_export['TIME'].value_counts()
```

```python
names = ['AUS']
coco.convert(names=names, to='name_short')
```

```python
import_export['Country'] = import_export['LOCATION'].apply(lambda x: coco.convert(names=[x], to='name_short'))
```

```python
import_export = import_export[import_export['Country'] != 'not found']
import_export = import_export[import_export['TIME'] >= 2019]
```

```python
import_export.columns = import_export.columns.str.title()
import_export.drop(columns=['Flag Codes'], inplace=True)
import_export
```
